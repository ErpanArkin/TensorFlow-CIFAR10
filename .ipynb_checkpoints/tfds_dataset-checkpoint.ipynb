```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds
%matplotlib inline
```

## Download and load the dataset

```python
ds, ds_info = tfds.load('cifar10', split='train', with_info=True)
label_names = ds_info.features["label"].names
print(ds_info)
```

## Show examples

```python
# hand-coded extration
def display_examples(nums=3):
    for i,example in enumerate(ds.take(nums)):
        print(label_names[example['label']], end=' ')
        plt.subplot(100+nums*10+i+1)
        plt.imshow(example['image'])
display_examples()
```

```python
# as pandas dataframe
tfds.as_dataframe(ds.take(4), ds_info)
```

```python
# using show_examples method
fig = tfds.show_examples(ds, ds_info)
```
